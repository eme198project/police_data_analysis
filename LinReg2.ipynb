```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm 
from scipy.stats import linregress 
from scipy import stats 
from statsmodels.sandbox.regression.predstd import wls_prediction_std
```

```python
df15 = pd.read_csv('UpdatedState2015.csv')
#adds column 'DeathsNorm', deaths per million in 2015
df15['DeathsNorm'] = (df15['Deaths']*(1e6)/df15['Population'])
df15state=df15[1:51]
df15state.at[5,'Mcdonalds']=1492
df15state.at[44,'Mcdonalds']=1224

df15state['Mcdonalds']=(pd.to_numeric(df15state['Mcdonalds'], errors='coerce'))
df15state
```

```python
# Regression Analysis for Population(predictor Variable) and Deaths (Response Variable) 
x = df15state['Population']
y = df15state['Deaths']
xMax = x.max()
xMin = x.min()
# start plot, called ax
fig, ax = plt.subplots(figsize = (8,4))
# plot the data points 
ax.scatter(x,y, alpha = 0.5)
# adds in grid for readability
ax.grid(True)
fig.suptitle('State Population and Deaths per State')
ax.set_xlabel("Population")
ax.set_ylabel("Deaths")
# constant intercept term to make the model y ~ x + c, not y~x
x = sm.add_constant(x)
# Model: y ~ x + c
model = sm.OLS(y,x)
fitted = model.fit()
# x values from xMin to xMax for plotting regression line
x_pred = np.linspace(xMin,xMax, 51)
# include intercept term to the predicting method 
x_pred2 = sm.add_constant(x_pred)
y_pred = fitted.predict(x_pred2)
# plot the regression line 
ax.plot(x_pred, y_pred, '-', linewidth =2)
## Confidence Interval 95%
y_hat = fitted.predict(x)
y_err = y - y_hat 
mean_x = x.T[1].mean()
n = len(x)
dof = n - fitted.df_model - 1
# t statistic 
t = stats.t.ppf(0.95, df = dof)
s_err = np.sum(np.power(y_err,2))
# confidence interval
conf = t * np.sqrt((s_err/(n-2))*(1.0/n + (np.power((x_pred-mean_x),2) /((np.sum(np.power(x_pred,2))) - n*(np.power(mean_x,2))))))
upper = y_pred + abs(conf)
lower = y_pred - abs(conf)
ax.fill_between(x_pred, lower,upper, color = '#888888', alpha = 0.4)
## Predictor Interval 
sdev, lower, upper = wls_prediction_std(fitted, exog = x_pred2, alpha = 0.05)
ax.fill_between(x_pred, lower, upper, color = '#888888', alpha = 0.1)

## Print Results 
linres = linregress(df15state['Population'], df15state['Deaths']) #x and y are arrays or lists
print(linres)
print(fitted.params)     # the estimated parameters for the regression line
print(fitted.summary())
```
